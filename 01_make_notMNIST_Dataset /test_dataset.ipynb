```
import dataset_practice as dp
import torch
from torch.utils.data import DataLoader
```

```
class LoadNotMNISTDataset():
    def __init__(self,
                 root:str,
                 download = False):
        self.train_dataset = dp.NotMNIST(root, train=True, download=download)
        self.test_dataset = dp.NotMNIST(root, train=False, download=download)

    def get_dataset(self, batch_size=100):
        train_iter = DataLoader(self.train_dataset, batch_size=batch_size, shuffle=True)
        test_iter = DataLoader(self.test_dataset, batch_size=batch_size, shuffle=True)

        return train_iter, test_iter
```
